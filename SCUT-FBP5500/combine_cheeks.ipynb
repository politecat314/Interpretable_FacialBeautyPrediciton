```python
import numpy as np
import cv2
import os
import matplotlib.pyplot as plt
```

```python
files = os.listdir("right_cheek")
files.remove("zero.npy")
```

```python
f = files[0]
image = cv2.imread(os.path.join("images",f))
left = cv2.imread(os.path.join("left_cheek",f))
right = cv2.imread(os.path.join("right_cheek",f))
if not os.path.exists("cheeks"):
    os.makedirs("cheeks")
```

```python
for f in files:
    left = cv2.imread(os.path.join("left_cheek",f))
    right = cv2.imread(os.path.join("right_cheek",f))

    if left.shape[0] > right.shape[0]:
        right = cv2.resize(right, (right.shape[1], left.shape[0]))

    elif right.shape[0] > left.shape[0]:
        left = cv2.resize(left, (left.shape[1], right.shape[0]))

    if left.shape[1] > right.shape[1]:
        right = cv2.resize(right, (left.shape[1], right.shape[0]))

    elif right.shape[1] > left.shape[1]:
        left = cv2.resize(left, (right.shape[1], left.shape[0]))

    cheek = np.hstack((left, right))
    cv2.imwrite(os.path.join("cheeks",f), cheek)
```
